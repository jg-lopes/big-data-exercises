````python

# We will convert the bigrams into key value pairs where the key is the bigram and the value is one (indicating this bigram has been found)

def map_bigrams(bigrams): 
  pairs = []
  for line in bigrams:
    pairs.append([(element, 1) for element in line])
  return pairs

mapped_vec = map_bigrams(bigrams)
print(mapped_vec[3])
````

````python

# We then set the keys of the key value pairs found in the map function as they keys of a hash table, with the values being the items of this table
# We can use a dict as a hash table in python

def shuffle_bigrams(map_vec):
  shuffle_dict = {}
  for line in map_vec:
    for key, value in line:
      shuffle_dict[key] = 1
  return shuffle_dict

shuffled_dict = shuffle_bigrams(mapped_vec)
print(list(shuffle_bigrams(mapped_vec).items())[0:50])
````

````python
# Since the assignment to 1 operation is both associative and commutative
# We can pre-compute it during the Map operation in order to reduce the number of bigrams we have to work with
# On this case, this means that if we have a duplicate of a single assignment in a Map node
# This map node will not send two of the same key value pairs for shuffling (which is redundant)
# Instead it will delete one of these pairs and send only one

# This reduces the amount of key value pairs that are being managed and processed, improving our performance and memory usage

# In order to do this, we can use a set instead of list in pair
# A set is an unordered collection of unique elements, thus it will remove duplicates
def map_combine_bigrams(bigrams): 
  pairs = []
  for line in bigrams:
    pairs.append({ (element, 1) for element in line })
  return pairs

# Comparing with combine using a simple example with duplicated bigrams
dupl_bigram = [[("A", "B"), ("A", "B"), ("B", "C")]]
print(f"Without combine: {map_bigrams(dupl_bigram)}")
print(f"With combine: {map_combine_bigrams(dupl_bigram)}")
````

````python
# We then set the keys of the key value pairs found in the map function as they keys of a hash table, with the values being the items of this table
# We can use a dict as a hash table in python
from collections import defaultdict

def shuffle_top_bigrams(map_vec): #to be completed
  shuffle_dict = defaultdict(list)
  for line in map_vec:
    for key, value in line:
      shuffle_dict[key].append(1)
  return shuffle_dict

shuffled_dict = shuffle_top_bigrams(mapped_vec)
print(list(shuffled_dict.items())[0:50])
````

````python
# For the final reduce we have two tasks to accomplish
# The first one is to sum all of the found binomes, in order to get the final amount of each binome
# Then, we need to extract the 5 highest binomes and return them to our user

# There are two approaches that can be done here
# 1: The first and more direct one is to implement a sorting algorithm value-wise in all pairs
# This is more general, and will solve the problem for the k-highest values
# However, this sorting has O(n log n) complexity

# 2: The second approach means using a heap priority queue 
# This method will create a heap and create pushes and pops on the k highest elements, in the measure 
# that new incoming elements from the list are higher than the current elements in the heap
# This results in a O(n log k) complexity -> O(n log 5) in our example
# Since k is smaller than n (normally much smaller), this approach is faster
# Drawback: need to define k a priori

import heapq
#Since we already have k = 5, we are going to apply the second method!
def reduce_top_bigrams(shuffle_dict):
  # Warning! This changes the dict by reference!
  # In order to avoid this we could create a copy of the dict
  # But in this example it would be unnecessary and only give extra memory usage
  for key, value in shuffle_dict.items():
    shuffle_dict[key] = sum(value)

  # APPROACH 1 - O (n log n)
  # Sorting value-wise and reversing the list so that it is descendent
  # result = sorted(shuffle_dict.items(), key=lambda kv: kv[1])[::-1]

  # APPROACH 2 - O (n log 5)
  number_of_elements = 5
  result = heapq.nlargest(number_of_elements, shuffle_dict.items(), key=lambda kv:kv[1])

  return result

result = reduce_top_bigrams(shuffled_dict)
print(result[0:5])
````

Describe and implement a Map-Reduce algorithm that, for each year-month, provides the top 10 products (the 10 products that have been sold the most):

2020-01: bread 82, milk 3, ...

````python
# Reading the file
# We apply a line by line read, removing the trailing line breaks
# We also split the individual comma separated elements into a list using the split function

def file_reader(filename):
    words = [line.strip().split(",") for line in open(filename)]
    return (words)

sm_lines = file_reader("supermarket.txt")
print(sm_lines)
````

As we can see, the first index of each line gives us the date in a yyyy-mm-dd format. The second to last element are the items.


````python
# We will create a function that will map the above lines into key-value pairs where
# key = tuple (date, product) where date and product are strings and date is yyyy-mm
# value = 1

# This will make so that our problem is modelled similarly to a word counting problem, except that we're counting the tuples
# Thus, we will group not only based on product, but on date as well

def map_top_products(lines):
  pairs = []
  for line in lines:

    # Splitting into [yyyy, mm, dd]
    date = line[0].split('-')
    # If mm has only one character, adding a zero before it
    if len(date[1]) == 1:
      date[1] = "0" + date[1]
    date = f"{date[0]}-{date[1]}"

    pairs.append([((date, element), 1) for element in line[1:]])
    
  return pairs

sm_mapped = map_top_products(sm_lines)
print(sm_mapped)
````

````python
# Similarly to the previous exercises, we use the defaultdict as a hash table for our keys, appending the values to a list
# But this time, since we have two nested keys, we are going to use a default dict of default dicts
# 2015-1 -> bread -> list
#        -> potatoes -> list
# 2015-2 -> etc.

# In a distributed system, this grouping has an advantage of splitting the months into indivudal nodes

def shuffle_top_products(mapped_list):
  shuffled_dict = defaultdict(lambda: defaultdict(list))
  for line in mapped_list:
    for key, value in line:
      date, product = key
      shuffled_dict[date][product].append(value)
  return shuffled_dict

sm_shuffled = shuffle_top_products(sm_mapped)
print(sm_shuffled)
````

````python
# The reduce will finally complete the final steps
# 1 -> Add up all of the data, product values
# 2 -> Extract the top products for each date
  # We use the same logic as the previous exercise on top binomes (with approach 2)
import heapq
def reduce_top_products(shuffled_dict):
  # Recall that ideally in a distributed system each reduce node would get a date
  # We can model the result for each one of the dates as a dict with key being the month
  # And the value the top products for the month
  result = {}
  for date, dict2 in shuffled_dict.items():
    
    for product, values in dict2.items():
      shuffled_dict[date][product] = sum(values)

    # Using heapq.nlargest to find the largets items
    number_of_elements = 10
    largest = heapq.nlargest(number_of_elements, shuffled_dict[date].items(), key=lambda kv:kv[1])
    result[date] = largest
  
  return result
  

sm_reduce = reduce_top_products(sm_shuffled)
print(sm_reduce)
````

````python
# All the available information is obtained in this function, however if we desire make the output as displayed we can do additional processing
# This can also be acoplated to the reduce function if the output is desired to be an output string instead of data

def print_top_products(reduced):
  # Optional: Sort by date the output (slower, but cleaner output) 
  for key in sorted(reduced):
    return_string = f"{key}: "
    for top_prods in reduced[key]:
      return_string = return_string + top_prods[0] + " " + str(top_prods[1]) + ", "
    print(return_string[:-2]) 

print_top_products(sm_reduce)
````

Describe and implement a Map-Reduce algorithm that gives, for each pair of products that appear in at least one ticket: the percentage of tickets in wich each product appears, the percentage of tickets in which the products appear together.

````python
# Reading the file
# We apply a line by line read, removing the trailing line breaks
# We also split the individual comma separated elements into a list using the split function

def file_reader(filename):
    words = [line.strip().split(",") for line in open(filename)]
    return (words)

sm_prices = file_reader("supermarket_prices.txt")
print(sm_prices)
````

````python
# Our expected end result will be a key value product where they key is either
# 1. The name of a single product (string) -> value then becomes the % of tickets with that product
# 2. A tuple (p1, p2) -> value is the % of the presence of the pair in a ticket

# NOTE: In order to avoid amiguity, this tuple is set so that p1 is always SMALLER than p2 
# Since pairing two products is commutative, the % of (p1,p2) is equal to the % of (p2,p1)
# In order to save computational resources and memory, we established by design that p1 < p2 
# (otherwise we would need to spend double the memory for each pair)
# Therefore, if p2 > p1, only the pair (p1,p2) exists as an output. It is equivalent to (p2,p1)

# Thus, in order to achieve this we will start with a map that takes our list and runs over it twice
# First, it extracts all products
# Then, it generates all combinations of products (pairs of purchases)
# Since we only care if a product was bought or not (and not its quantity), we remove duplicates

# The result of the map is a k, v pair where k = product or (p1,p2) tuple and v = 1 indicating the key was in a ticket

# Finally, we have a special key = "tickets", with v = 1 indicating that we processed a ticket (to be used for percentages!)
import itertools
def map_avgper(lines):
  pairs = []
  for line in lines:

    # Lists the single product purchases
    products = line[1:]

    #Removing the unused price
    for i, product in enumerate(products):
      products[i] = product.split("-")[0]
    # Gets all possible pairs in list
    pairwise_prods = list(itertools.combinations(products, 2))
    
    single_purchases = [((element), 1) for element in line[1:]]
    # Lists the pairs, sorts them so to avoid ambiguity (some associations to (p1,p2) and others to (p2,p1) when they are equivalent)
    map_pairwise = [(tuple(sorted(element)), 1) for element in pairwise_prods]
    
    # We don't want to double count when a person buys twice of a product
    # Therefore, we create a set to force uniqueness, before reconverting to a list
    
    single_purchases = list(set(single_purchases))
    map_pairwise = list(set(map_pairwise))

    # Tickets counts the amount of tickets
    pairs.append(map_pairwise + single_purchases + [("tickets", 1)])
    

  return pairs

sm_mapped = map_avgper(sm_lines)
for line in sm_mapped[:3]:
  print(line)
````

````python
# Grouping by key
def shuffle_avgper(mapped):
  groups = defaultdict(list)
  for line in mapped:
    for key, value in line:
      groups[key].append(value)
  return groups

sm_shuffled = shuffle_avgper(sm_mapped)
print(sm_shuffled)
````

````python
# Adding all the different number of tickets they key elements were into

# As a last step, we add all elements of tickets in order to find the total number of tickets
# Finally, we divide our number of tickets found for the key by the number of total keys to find the percentage
def reduce_avgper(shuffled):
  # Get absolute quantity of items
  for key, value in shuffled.items():
    shuffled[key] = sum(value)
  
  # Average it over the amount of tickets
  # In a distributed system, this value would need to be communicated from the node that computes it to all the others
  # (or it would be computed in every single node)
  # Not efficient! Can be improved with combine
  for key, value in shuffled.items():
    if key != "tickets":
      shuffled[key] = shuffled[key] / shuffled["tickets"]
  return shuffled

sm_reduced = reduce_avgper(sm_shuffled)

print(sm_reduced)
````

````python
# We already partially the combine by forcing uniqueness in our data!
# However we can eagerly compute the value of tickets so that we divide it on the assignment of the value for our keys
# This works because division is commutative and associative, and the number of tickets is constant

def map_combine_avgper(lines):
  pairs = []
  
  # Could be done with len(lines), only doing this explictly to emphasize the counting aspect
  tickets = 0
  for line in lines:
    tickets += 1
  
  for line in lines:
    # Lists the single product purchases
    products = line[1:]
    # Gets all possible pairs in list
    pairwise_prods = list(itertools.combinations(products, 2))
    
    single_purchases = [((element), 1/tickets) for element in line[1:]]
    # Lists the pairs, sorts them so to avoid ambiguity (some associations to (p1,p2) and others to (p2,p1) when they are equivalent)
    map_pairwise = [(tuple(sorted(element)), 1/tickets) for element in pairwise_prods]
    
    # We don't want to double count when a person buys twice of a product
    # Therefore, we create a set to force uniqueness, before reconverting to a list
    
    single_purchases = list(set(single_purchases))
    map_pairwise = list(set(map_pairwise))

    # Tickets counts the amount of tickets
    pairs.append(map_pairwise + single_purchases)
    

  return pairs

sm_mapped = map_combine_avgper(sm_lines)
for line in sm_mapped[:3]:
  print(line)
````
